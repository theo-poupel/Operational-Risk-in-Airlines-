```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings

warnings.filterwarnings("ignore")

month_year = [
    "Nov-18","Oct-18","Sep-18","Aug-18","Jul-18","Jun-18","May-18","Apr-18","Mar-18","Feb-18","Jan-18","Dec-17",
    "Nov-17","Oct-17","Sep-17","Aug-17","Jul-17","Jun-17","May-17","Apr-17","Mar-17","Feb-17","Jan-17","Dec-16",
    "Nov-16","Oct-16","Sep-16","Aug-16","Jul-16","Jun-16","May-16","Apr-16","Mar-16","Feb-16","Jan-16","Dec-15",
    "Nov-15","Oct-15","Sep-15","Aug-15","Jul-15","Jun-15","May-15","Apr-15","Mar-15","Feb-15","Jan-15","Dec-14"
]

total_delay_cost = [
    146720.00,  40640.00,  65360.00, 225200.00, 256960.00, 188320.00,
    112640.00, 105760.00, 103920.00, 109120.00, 242160.00, 249360.00,
     73360.00,  40480.00,  65920.00, 106240.00, 155600.00, 120400.00,
     74720.00, 326320.00, 132160.00, 173840.00, 301840.00, 265840.00,
     60880.00,  10880.00,  62320.00, 103120.00, 207120.00, 159520.00,
     60480.00,  45680.00,  68320.00, 107120.00, 159840.00, 167440.00,
     76000.00,  27680.00,  66160.00, 115360.00, 149680.00, 150160.00,
    119200.00,  98960.00, 174400.00, 326000.00, 175280.00,  78160.00
]

# Convert to DataFrame
df = pd.DataFrame({
    "month_year": month_year,
    "cost_raw": total_delay_cost
})

df["ds"] = pd.to_datetime(df["month_year"], format="%b-%y")
df["y"] = df["cost_raw"] / 1_000_000  # convert to millions

# Sort oldest → newest
df.sort_values("ds", inplace=True)
df.reset_index(drop=True, inplace=True)

df["trend"] = np.arange(len(df))
df["month"] = df["ds"].dt.month

X = pd.DataFrame({"const": 1, "trend": df["trend"]})
for m in range(1, 12):  # month 1–11, month 12 baseline
    X[f"month_{m}"] = (df["month"] == m).astype(int)

y = df["y"].values

beta, *_ = np.linalg.lstsq(X.values, y, rcond=None)

# Residuals for CI
resid = y - (X.values @ beta)
resid_std = resid.std()

future_periods = 36
future_dates = pd.date_range(start=df["ds"].max() + pd.DateOffset(months=1),
                             periods=future_periods, freq="MS")

df_future = pd.DataFrame({
    "ds": future_dates,
    "trend": np.arange(len(df), len(df) + future_periods)
})
df_future["month"] = df_future["ds"].dt.month

X_future = pd.DataFrame({"const": 1, "trend": df_future["trend"]})
for m in range(1, 12):
    X_future[f"month_{m}"] = (df_future["month"] == m).astype(int)

baseline_forecast = X_future.values @ beta

# Mimic Prophet:
# - Lower band ~ 95% CI  → multiplier ≈ -2σ
# - Upper band ~ 80% CI  → multiplier ≈ +1.28σ

low_band = baseline_forecast - 2.0 * resid_std
high_band = baseline_forecast + 1.28 * resid_std

plt.figure(figsize=(15, 9))
ax = plt.gca()

# Historical
plt.plot(df["ds"], df["y"],
         label="Historical Data",
         color="black", linewidth=2)

# Baseline (medium)
plt.plot(df_future["ds"], baseline_forecast,
         label="Medium Scenario (Baseline Forecast)",
         color="green", linestyle="--", linewidth=2)

# Low scenario
plt.plot(df_future["ds"], low_band,
         label="Low Scenario (95% Lower Bound)",
         color="blue", linewidth=2)

# High scenario
plt.plot(df_future["ds"], high_band,
         label="High Scenario (80% Upper Bound)",
         color="red", linewidth=2)

# Hybrid fill
plt.fill_between(df_future["ds"], low_band, high_band,
                 color="gray", alpha=0.2)

# Formatting
ax.axvline(x=df["ds"].max(), color="black", linestyle=":")
ax.set_title("LAX Hybrid Scenario Analysis (Asymmetric Statistical Intervals)",
             fontsize=16)
ax.set_ylabel("Total Monthly Delay Cost ($ Millions)", fontsize=12)
ax.set_xlabel("Date", fontsize=12)
plt.legend(loc="upper left")
plt.grid(True, alpha=0.3)

plt.xlim([pd.to_datetime("2015-01-01"), df_future["ds"].max()])
ax.xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 7]))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%b-%y"))
plt.setp(ax.get_xticklabels(), rotation=45, ha="right")

plt.tight_layout()
plt.savefig("lax_hybrid_scenarios_no_prophet.png")
plt.show()
```
